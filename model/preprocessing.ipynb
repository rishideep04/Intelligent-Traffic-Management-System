```python
import os

# Specify the folder paths
images_folder = "train/images"
labels_folder = "train/labels"

# Get the list of images and labels
images = sorted([f for f in os.listdir(images_folder) if f.endswith(('.jpg', '.png', '.jpeg'))])
labels = sorted([f for f in os.listdir(labels_folder) if f.endswith('.txt')])

# Ensure the folders have the same number of files
if len(images) != len(labels):
    print("The number of images and labels do not match. Check the folders.")
    exit()

# Rename the files
for i, (image, label) in enumerate(zip(images, labels), start=1):
    new_image_name = f"{i}.jpg"
    new_label_name = f"{i}.txt"

    # Full paths for renaming
    image_path = os.path.join(images_folder, image)
    new_image_path = os.path.join(images_folder, new_image_name)
    label_path = os.path.join(labels_folder, label)
    new_label_path = os.path.join(labels_folder, new_label_name)

    # Rename the files
    os.rename(image_path, new_image_path)
    os.rename(label_path, new_label_path)

print("Renaming completed!")
```

```python
import cv2
import os
import random
import matplotlib.pyplot as plt

# Define paths
image_folder = "images/"
label_folder = "labels/"
output_folder = "output/"  # Optional: Save images with bounding boxes

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Get a list of image files
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Randomly select 10 images
random_images = random.sample(image_files, 10)

# Function to visualize bounding boxes
def visualize_bounding_boxes(image_path, label_path, save_path=None):
    # Read the image
    image = cv2.imread(image_path)
    height, width, _ = image.shape

    # Read the corresponding label file
    with open(label_path, 'r') as f:
        lines = f.readlines()

    for line in lines:
        # Parse label data (assuming YOLO format: class x_center y_center width height)
        parts = line.strip().split()
        class_id = int(parts[0])
        x_center, y_center, box_width, box_height = map(float, parts[1:])

        # Convert normalized coordinates to pixel values
        x_min = int((x_center - box_width / 2) * width)
        y_min = int((y_center - box_height / 2) * height)
        x_max = int((x_center + box_width / 2) * width)
        y_max = int((y_center + box_height / 2) * height)

        # Draw the bounding box
        cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
        cv2.putText(image, f"Class {class_id}", (x_min, y_min - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Save or display the image
    if save_path:
        cv2.imwrite(save_path, image)
    else:
        # Display the image using matplotlib
        plt.figure(figsize=(10, 10))
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.show()

# Process only 10 random images
for image_name in random_images:
    image_path = os.path.join(image_folder, image_name)
    label_path = os.path.join(label_folder, image_name.replace('.jpg', '.txt').replace('.png', '.txt').replace('.jpeg', '.txt'))

    if os.path.exists(label_path):
        output_path = os.path.join(output_folder, image_name)
        visualize_bounding_boxes(image_path, label_path, save_path=output_path)
        print(f"Processed {image_name}")
    else:
        print(f"Label file missing for {image_name}")
```
