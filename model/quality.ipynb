```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageStat

def calculate_brightness(image_path):
    """Calculate brightness as the mean pixel value of the grayscale image."""
    img = Image.open(image_path).convert('L')  # Convert to grayscale
    stat = ImageStat.Stat(img)
    return stat.mean[0]  # Mean brightness

def calculate_sharpness(image_path):
    """Calculate sharpness using the variance of the Laplacian method."""
    img_cv = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Read in grayscale
    laplacian_var = cv2.Laplacian(img_cv, cv2.CV_64F).var()
    return laplacian_var  # Higher values indicate sharper images

# Set your image path here
image_path = "train/images/ulu2832_jpg.rf.c9e52329d83f53e7eb90f445338a37d0.jpg"  # Change this to your image path

# Load the image using PIL
image = Image.open(image_path)

# Display the image using Matplotlib
plt.figure(figsize=(6,6))
plt.imshow(image)
plt.axis('off')
plt.title("Uploaded Image")
plt.show()

# Calculate brightness and sharpness
brightness = calculate_brightness(image_path)
sharpness = calculate_sharpness(image_path)

# Print results
print(f"Brightness: {brightness:.2f}")
print(f"Sharpness: {sharpness:.2f}")
```
